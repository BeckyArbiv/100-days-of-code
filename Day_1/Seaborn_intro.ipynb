```python
import numpy as np
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
data_filepath = '~/Downloads/cancer_b.csv'
cancer_data = pd.read_csv(data_filepath, index_col="Id")
cancer_data.head()
```

```python
plt.figure(figsize=(12,5))
sns.lineplot(data=cancer_data['Radius (mean)'],)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=cancer_data.index, y=cancer_data['Radius (mean)'])
```

```python
plt.figure(figsize=(14,7))
sns.heatmap(data=cancer_data.iloc[0:10,1:10], annot=True)

```
